```python
import pandas as pd
import matplotlib.pyplot as plt
from btgenerate.database.dataloader import LiquidMasterTableDataSet

# pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
pd.set_option('display.colheader_justify', 'center')
pd.set_option('display.precision', 3)
```

```python
ds_lmt = LiquidMasterTableDataSet()
ds_lmt.normalize_components(inplace=True)
ds_lmt.chemicals
```

```python
group_ids = []
i_iter = 0
num_recipes = len(ds_lmt.chemicals)
chemical_presence = (ds_lmt.chemicals != 0)
for id in chemical_presence.index:
    i_iter += 1
    if len(group_ids) == 0:
        group_ids.append([id])
        continue
    for group in group_ids:
        if not (chemical_presence.loc[id] ^ chemical_presence.loc[group[0]]).any(): # bitwise XOR -> group and row have same components, add to the group
            group.append(id)
            break
    else:
        group_ids.append([id]) # not group has same component as row, create a new group     
    print(f"Data compared {i_iter}/{num_recipes}", end="\r", flush=True)
```

```python
group_dfs = []
for group in group_ids:
    chemicals = ds_lmt.chemicals.loc[group]
    absent_chemicals = chemicals.columns[(chemicals == 0).all(axis=0)]
    df = pd.concat(
        [
            ds_lmt.electrolyte_ids.loc[group],
            ds_lmt.chemicals.loc[group, ~chemicals.columns.isin(absent_chemicals)],
            ds_lmt.lce[group]
        ], axis=1
    )
    group_dfs.append(df)
```

### num_chemicals: 2

```python
num_compoment_limit = 2
i = 1
for df in group_dfs:
    if (df.shape[1] == num_compoment_limit + 2) and (df.shape[0] > 1):
        plt.figure(figsize=(df.shape[1]-1, 4))
        pd.plotting.parallel_coordinates(df, class_column="Electrolyte ID")
        plt.title(f"{num_compoment_limit}.{i}")
        i += 1
        plt.show()
```
